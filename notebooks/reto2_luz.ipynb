# Reto 2 – Análisis de luz

**Objetivo:** Explorar cómo la luz influye sobre la biomasa y otros parámetros.

Carga el conjunto de datos y explora las variables relevantes.

```python
import pandas as pd

df = pd.read_csv("../data/resultados_experimentos.csv")
df.head(20)
```

```python
def indice_luz(df):
    # Calcula la luz máxima del conjunto
    luz_max = df["luz"].max()

    # Retorna la razón entre cada valor de luz y la luz máxima
    return df["luz"] / luz_max
```

```python
df["indice_luz"] = indice_luz(df)
```

```python
p33 = df["luz"].quantile(0.33)
p66 = df["luz"].quantile(0.66)
```

```python

def clasificar_luz(valor):
    if valor <= p33:
        return "baja"
    elif valor <= p66:
        return "media"
    else:
        return "alta"
```

```python
import matplotlib.pyplot as plt

# Calcular biomasa promedio por categoría
biomasa_prom = df.groupby("categoria_luz")["biomasa"].mean()

# Crear gráfico de barras
plt.figure(figsize=(8,5))
plt.bar(biomasa_prom.index, biomasa_prom.values)
plt.xlabel("Categoría de luz")
plt.ylabel("Biomasa promedio")
plt.title("Biomasa promedio por categoría de luz")
plt.show()
```
