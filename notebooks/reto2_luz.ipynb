# Reto 2 – Análisis de luz

**Objetivo:** Explorar cómo la luz influye sobre la biomasa y otros parámetros.

Carga el conjunto de datos y explora las variables relevantes.

```python
import pandas as pd

df = pd.read_csv("../data/resultados_experimentos.csv")
df.head(20)
```

```python
1.
def indice_luz(df):
    """
    Calcula el índice de luz para cada sitio.
    El índice de luz es la razón entre la luz de cada sitio y la luz máxima del conjunto.

    Args:
        df (pd.DataFrame): El DataFrame de entrada que contiene la columna 'luz'.

    Returns:
        pd.Series: Una Serie que contiene el índice de luz para cada fila.
    """
    max_luz = df['luz'].max()
    df['indice_luz'] = df['luz'] / max_luz
    return df
```

```python
df = indice_luz(df)
display(df.head(20))
```

```python
2.
display(df.head(20))
```

```python
3.
# Calcular los percentiles 33 y 66 de la columna 'luz'
percentil_33 = df['luz'].quantile(0.33)
percentil_66 = df['luz'].quantile(0.66)

print(f"Percentil 33 de 'luz': {percentil_33:.2f}")
print(f"Percentil 66 de 'luz': {percentil_66:.2f}")

def clasificar_luz(valor):
    """
    Clasifica un valor de luz en 'baja', 'media' o 'alta' según percentiles.

    Args:
        valor (float): El valor de luz a clasificar.

    Returns:
        str: La clasificación del valor de luz ('baja', 'media' o 'alta').
    """
    if valor <= percentil_33:
        return 'baja'
    elif valor <= percentil_66:
        return 'media'
    else:
        return 'alta'

# Aplicar la función al DataFrame para crear una nueva columna 'clase_luz'
df['clase_luz'] = df['luz'].apply(clasificar_luz)

display(df.head(20))
```

```python
4.
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular la biomasa promedio por clase de luz
biomasa_promedio_por_clase_luz = df.groupby('clase_luz')['biomasa'].mean().reset_index()

# Definir el orden de las categorías para el gráfico
orden_clase_luz = ['baja', 'media', 'alta']
biomasa_promedio_por_clase_luz['clase_luz'] = pd.Categorical(biomasa_promedio_por_clase_luz['clase_luz'], categories=orden_clase_luz, ordered=True)
biomasa_promedio_por_clase_luz = biomasa_promedio_por_clase_luz.sort_values('clase_luz')

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
sns.barplot(x='clase_luz', y='biomasa', data=biomasa_promedio_por_clase_luz, palette='viridis')
plt.title('Biomasa Promedio por Categoría de Luz')
plt.xlabel('Clase de Luz')
plt.ylabel('Biomasa Promedio')
plt.show()
```
